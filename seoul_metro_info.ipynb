```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('seoul-metro-2021.logs.csv')
data
```

```python
station_info = pd.read_csv('seoul-metro-station-info.csv')
station_info
```

```python
data.info()
```

```python
station_info.info()
```

```python
station_sum = data.groupby('station_code').sum()
station_sum
# new_data = data.groupby('station_code').agg({'people_in': 'sum', 'people_out': 'sum'})
# new_data
```

```python
station_info = station_info[['station.code', 'geo.latitude', 'geo.longitude']]
station_info
```

```python
station_info.info()
```

```python
station_info = station_info.set_index('station.code')
station_info
```

```python
join_data = station_sum.join(station_info)
join_data
```

```python
seoul_in = folium.Map(location=[37.55, 126.98], zoom_start = 12)
seoul_in
```

```python
from folium.plugins import HeatMap

HeatMap(data = join_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(seoul_in)
seoul_in
```

```python
seoul_out = folium.Map(location=[37.55, 126.98], zoom_start = 12)

HeatMap(data = join_data[['geo.latitude', 'geo.longitude', 'people_out']]).add_to(seoul_out)
seoul_out
```

```python
# 출근용
morning_data = data[pd.to_datetime(data.timestamp).dt.hour < 9]
morning_data
```

```python
# 퇴근용
evening_data = data[pd.to_datetime(data.timestamp).dt.hour > 17]
evening_data
```

```python
# 역 기준으로 그룹화하여 데이터 더하기
morning_station_sum = morning_data.groupby("station_code").sum()
evening_station_sum = evening_data.groupby("station_code").sum()
```

```python
# 역 정보 합치기
morning_joined_data = morning_station_sum.join(station_info)
evening_joined_data = evening_station_sum.join(station_info)
```

```python
# 출근 시간 승차 인원
morning_seoul_in = folium.Map(location=[37.55, 126.98], zoom_start = 12)
HeatMap(data = morning_joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(morning_seoul_in)
morning_seoul_in
```

```python
# 출근 시간 하차 인원
morning_seoul_out = folium.Map(location=[37.55, 126.98], zoom_start = 12)
HeatMap(data = morning_joined_data[['geo.latitude', 'geo.longitude', 'people_out']]).add_to(morning_seoul_out)
morning_seoul_out
```

```python
# 퇴근 시간 승차 인원
evening_seoul_in = folium.Map(location=[37.55, 126.98], zoom_start = 12)
HeatMap(data = evening_joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(evening_seoul_in)
evening_seoul_in
```

```python
# 퇴근 시간 하차 인원
evening_seoul_out = folium.Map(location=[37.55, 126.98], zoom_start = 12)
HeatMap(data = evening_joined_data[['geo.latitude', 'geo.longitude', 'people_out']]).add_to(evening_seoul_out)
evening_seoul_out
```

```python
lunch_data =  data[(pd.to_datetime(data.timestamp).dt.hour >= 11) & (pd.to_datetime(data.timestamp).dt.hour <= 14)]
lunch_data
```

```python
lunch_station_sum = lunch_data.groupby('station_code').sum()
```

```python
lunch_joined_data = lunch_station_sum.join(station_info)
lunch_joined_data
```

```python
lunch_seoul_in = folium.Map(location=[37.55, 126.98], zoom_start = 12)
HeatMap(data = lunch_joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(lunch_seoul_in)
lunch_seoul_in
```

```python
lunch_seoul_out = folium.Map(location=[37.55, 126.98], zoom_start = 12)
HeatMap(data = lunch_joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(lunch_seoul_out)
lunch_seoul_out
```

```python
lunch_seoul_out_circle = folium.Map(location=[37.55, 126.98], zoom_start = 12)

for index, row in lunch_joined_data[['geo.latitude', 'geo.longitude', 'people_out']].iterrows():
    if int(row['people_out']) > 2_500_000:
        folium.Circle(location=[str(row['geo.latitude']), str(row['geo.longitude'])],   color='Red', fill=True,
              radius=(int(row['people_out'])/10000)).add_to(lunch_seoul_out_circle)

lunch_seoul_out_circle


# lunch_seoul_out = folium.Map(location=[37.55, 126.98], zoom_start=12)

# filtered_data = lunch_joined_data[lunch_joined_data['people_out'] > 2500000]

# for index, row in filtered_data.iterrows():
#     radius = 0.0001 * row['people_out']
#     folium.Circle(location=[row['geo.latitude'], row['geo.longitude']],
#                  radius=radius,
#                  color='red',
#                  fill=True,
#                  fill_color='red',
#                  fill_opacity=0.3
#                  ).add_to(lunch_seoul_out)

# lunch_seoul_out
```
